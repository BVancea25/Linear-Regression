```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from six.moves import urllib
import tensorflow as tf
from tensorflow import feature_column as fc
```

```python
dftrain = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/train.csv') # training data
dfeval = pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/eval.csv') # testing data
y_train = dftrain.pop('survived')
y_eval = dfeval.pop('survived')
```

```python
categorical=['sex','n_siblings_spouses','parch','class','deck','embark_town','alone']
numeric=['age','fare']
```

```python
feature_columns=[]
for feature_name in categorical:
    vocabulary=dftrain[feature_name].unique()
    feature_columns.append(tf.feature_column.categorical_column_with_vocabulary_list(feature_name,vocabulary))
for feature_name in numeric:
    feature_columns.append(tf.feature_column.numeric_column(feature_name,dtype=tf.float32))
```

```python
print(feature_columns)
```

```python
def make_input_fn(data_df,label_df,epochs=10,shuffle=True,batch_number=32):
    def input_function():
        ds=tf.data.Dataset.from_tensor_slices((dict(data_df),label_df))
        if shuffle:
            ds=ds.shuffle(1000)
        ds=ds.batch(batch_number).repeat(epochs)
        return ds
    return input_function
```

```python
train_input_fn=make_input_fn(dftrain,y_train)
eval_input_fn=make_input_fn(dfeval,y_eval,epochs=1,shuffle=False)
```

```python
linear_est=tf.estimator.LinearClassifier(feature_columns=feature_columns)
linear_est.train(train_input_fn)
result=linear_est.evaluate(eval_input_fn)

print(result['accuracy'])
```

```python
result=list(linear_est.predict(eval_input_fn))
print(dfeval.loc[0])
print(result[0]['probabilities'][1])
```
